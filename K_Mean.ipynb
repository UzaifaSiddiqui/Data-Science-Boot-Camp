```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df.drop(["petal length (cm)","petal width (cm)"],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

```python
labels = kmeans.fit_predict(df)
```

```python
labels
```

```python
kmeans.cluster_centers_
```

```python
df.head()
```

```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='red')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
selected_columns = ['sepal length (cm)','sepal width (cm)']
data_to_scale = df[selected_columns]

scaled_data = scaler.fit_transform(data_to_scale)

df[selected_columns] = scaled_data
```

```python
df
```

```python
labels2 = kmeans.fit_predict(df)
```

```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=labels2, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='red')
plt.show()
```

```python
from sklearn.metrics import silhouette_score


# Assuming 'data' is your dataset

# List to store silhouette scores
silhouette_scores = []

# Range of k values to try
k_range = range(2, 11)  # Try different values of k from 2 to 10

# Iterate over each value of k
for k in k_range:
    # Create KMeans instance with k clusters
    kmeans = KMeans(n_clusters=k, random_state=42)

    # Fit the model to your data
    kmeans.fit(df)

    # Compute the silhouette score
    silhouette_avg = silhouette_score(df, kmeans.labels_)

    # Append the silhouette score to the list
    silhouette_scores.append(silhouette_avg)

# Plotting the silhouette scores
plt.plot(k_range, silhouette_scores, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score For Optimal k')
```
