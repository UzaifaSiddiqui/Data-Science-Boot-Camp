```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("train.csv")
df1
```

```python

df2 = pd.read_csv("test.csv")
df2
```

```python
df = pd.concat([df1,df2],ignore_index=True)
df
```

```python
df = df.dropna(subset=['Survived'])
df = df.dropna(subset=['Embarked'])
```

```python
df.info()
```

```python
df.drop(['Name','Cabin'],axis='columns',inplace=True)
```

```python
df[df.duplicated() == True]
```

```python
rounding_off = df["Fare"]
new_fare = [int(i) for i in rounding_off]
df['Fare']=new_fare
```

```python
for column in df.columns:
  try:
    plt.figure(figsize=(8, 5))
    plt.hist(df[column], bins=8, color='skyblue', edgecolor='black')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title('Histogram of ' + column)
    plt.grid(True)
    plt.show()
  except Exception as e:
    print(f"Can not be plotted for {column} column")
```

```python
for column in df.columns:
  if df[column].dtype !="object":
    plt.scatter(df["Survived"], df[column])
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Scatter Plot Example')
    plt.show()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="mean")
null=df.isnull().sum()
print([null])
```

```python
imputer.fit(df[["Age"]])
df["Age"]=imputer.transform(df[["Age"]])
```

```python
df.info()
```

```python
df_encoded  = pd.get_dummies(df, columns=['PassengerId','Sex','Ticket','Embarked'])
```

```python
df_encoded
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = df_encoded.drop(['Survived'], axis="columns")
y = df_encoded['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted=model.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , y_predicted)
sns.heatmap(cm , annot=True)
plt.xlabel("Truth")
plt.ylabel("Predicted")
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score
```

```python
print ('Accuracy:', accuracy_score(y_test, y_predicted))
precisions = cross_val_score(model, X_train, y_train, cv=5,
scoring='precision')
print ('Precision', np.mean(precisions), precisions)
recalls = cross_val_score(model, X_train, y_train, cv=5,
scoring='recall')
print ('Recalls', np.mean(recalls), recalls)
f1_score = cross_val_score(model, X_train, y_train, cv=5,
scoring='f1')
print ('F1', np.mean(f1_score), f1_score)
```
