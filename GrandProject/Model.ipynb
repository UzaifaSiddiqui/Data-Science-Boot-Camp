```python
import pandas as pd
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("main.csv")
data
```

```python
data.drop("Unnamed: 0",axis="columns",inplace=True)
```

```python
X = data.drop(['Status'], axis="columns")
y = data['Status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_test.info()
```

```python
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(X_train,y_train)
```

```python
def make_prediction(age, income, emp_length, amount, rate, status, percent_income, cred_length, home_mortgage, home_other, home_own, home_rent, intent_debtconsolidation, intent_education, intent_homeimprovement, intent_medical, intent_personal, intent_venture, default_n, default_y):
    try:
        # Convert inputs to the correct types, replacing None with 0
        age = int(age) if age is not None else 0
        income = int(income) if income is not None else 0
        emp_length = float(emp_length) if emp_length is not None else 0
        amount = int(amount) if amount is not None else 0
        rate = float(rate) if rate is not None else 0
        percent_income = float(percent_income) if percent_income is not None else 0
        cred_length = int(cred_length) if cred_length is not None else 0
        home_mortgage = int(home_mortgage) if home_mortgage is not None else 0
        home_other = int(home_other) if home_other is not None else 0
        home_own = int(home_own) if home_own is not None else 0
        home_rent = int(home_rent) if home_rent is not None else 0
        intent_debtconsolidation = int(intent_debtconsolidation) if intent_debtconsolidation is not None else 0
        intent_education = int(intent_education) if intent_education is not None else 0
        intent_homeimprovement = int(intent_homeimprovement) if intent_homeimprovement is not None else 0
        intent_medical = int(intent_medical) if intent_medical is not None else 0
        intent_personal = int(intent_personal) if intent_personal is not None else 0
        intent_venture = int(intent_venture) if intent_venture is not None else 0
        default_n = int(default_n) if default_n is not None else 0
        default_y = int(default_y) if default_y is not None else 0

        # Load the machine learning model
        with open("loan_predictor.pkl", "rb") as f:
            model = pickle.load(f)

        # Predict using the loaded model and input features
        preds = model.predict([[age, income, emp_length, amount, rate, percent_income, cred_length, home_mortgage, home_other, home_own, home_rent, intent_debtconsolidation, intent_education, intent_homeimprovement, intent_medical, intent_personal, intent_venture, default_n, default_y]])

        if preds == 1:
            return "You are eligible for the loan"
        else:
            return "You are not eligible for the loan"

    except (ValueError, TypeError) as e:
        return f"Error: {e}. Please ensure all input values are in the correct format."

# Define input components
age_input = gr.Number(label="Enter your age")
income_input = gr.Number(label="Enter your income")
emp_length_input = gr.Number(label="Enter your employment length")
amount_input = gr.Number(label="Enter the amount")
rate_input = gr.Number(label="Enter the rate")
percent_income_input = gr.Number(label="Enter the percentage of your income")
cred_length_input = gr.Number(label="Enter your credit length")
home_mortgage_input = gr.Checkbox(label="Select if you have a mortgage")
home_other_input = gr.Checkbox(label="Select if you have other home ownership")
home_own_input = gr.Checkbox(label="Select if you own a home")
home_rent_input = gr.Checkbox(label= "Select if you rent a home")
intent_debtconsolidation_input = gr.Checkbox(label="Select if your intent is debt consolidation")
intent_education_input = gr.Checkbox(label="Select if your intent is education")
intent_homeimprovement_input = gr.Checkbox(label="Select if your intent is home improvement")
intent_medical_input = gr.Checkbox(label="Select if your intent is medical")
intent_personal_input = gr.Checkbox(label="Select if your intent is personal")
intent_venture_input = gr.Checkbox(label="Select if your intent is venture")
default_n_input = gr.Checkbox(label="Select if any previous default history is no")
default_y_input = gr.Checkbox(label="Select if donot have any previous history default is yes")

# Define the output
output = gr.Textbox(label="Output")

# Create the Gradio interface
app = gr.Interface(fn=make_prediction,
                   inputs=[age_input, income_input, emp_length_input, amount_input, rate_input, percent_income_input, cred_length_input,
                           home_mortgage_input, home_other_input, home_own_input, home_rent_input, intent_debtconsolidation_input, intent_education_input,
                           intent_homeimprovement_input, intent_medical_input, intent_personal_input, intent_venture_input, default_n_input, default_y_input],
                   outputs=output,title="Loan Eligibility Predictor")

# Launch the interface
app.launch(debug=True)
```
