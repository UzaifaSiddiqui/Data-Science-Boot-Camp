```python
for column in df.columns:
    plt.figure(figsize=(8, 5))
    plt.hist(df[column], bins=8, color='skyblue', edgecolor='black')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title('Histogram of ' + column)
    plt.grid(True)
    plt.show()
```

```python
sns.heatmap(df.corr(),cmap="viridis",annot=True)
plt.title("Correlation Matrix")
```

```python
object_type = [column for column in df.columns if df[column].dtype == 'object']
object_type
```

```python
df_encoded  = pd.get_dummies(df, columns=object_type)
df_encoded
```

```python
for column in df.columns:
    plt.figure(figsize=(8, 6))
    plt.scatter(df['Total Marks'],df[column], color='blue', alpha=0.5)
    plt.xlabel('Total Marks')
    plt.ylabel(column)
    plt.title('Scatter Plot of ' + 'Total Marks' + ' vs ' + column)
    plt.grid(True)
    plt.show()
```
